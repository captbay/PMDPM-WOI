```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X = df1.drop(['id','name','est_diameter_max','orbiting_body','sentry_object','hazardous'],axis=1)
y = df1.hazardous.astype('int')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_mmscaled = scaler.transform(X_train)
X_test_mmscaled = scaler.transform(X_test)

# print("Score train before scaled: ".format(scaler.score(X_train)))
# print("Score test before scaled: ".format(scaler.score(X_test)))
# print("Score train after scaled: ".format(scaler.score(X_train_mmscaler)))
# print("Score test after scaled: ".format(scaler.score(X_test_mmscaler)))

print("X_train_mmscaled : ")
print(X_train_mmscaled)
print("\n\nX_test_mmscaled : ")
print(X_test_mmscaled)

print("\n\n\nper-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(
    X_train_mmscaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    X_train_mmscaled.max(axis=0)))
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

X_train_stdscaled = std_scaler.fit(X_train).transform(X_train)
X_test_stdscaled = std_scaler.transform(X_test)

print('X_train Standard Scaler: ')
print(X_train_stdscaled)
print('\n\nX_test Standard Scaler: ')
print(X_test_stdscaled)

print("\n\n\nper-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))

print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(
    X_train_stdscaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    X_train_stdscaled.max(axis=0)))
```
